```python
import os,sys
from scipy.stats.stats import pearsonr
from scipy.cluster.vq import kmeans, vq
import numpy as np

data_file = "output.contigs2reads.contigs.short.txt"

data = {}
with open(data_file, 'r') as fin:
    for l in fin:
        p=l.strip().split("\t")
        did = p.pop(0)
        p = map(float, p)
        data[did] = p

# now we need to find all pairwise correlation scores
ids = data.keys()
```

```python
len(ids)
```

```python
pairwise =[]
for i in ids:
    temp = []
    for j in ids:
        temp.append(0)
    pairwise.append(temp)

# set ourselves to 1
for i in range(len(ids)):
    pairwise[i][i] = 1

# measure all correlations
for i in range(len(ids)):
    for j in range(i+1, len(ids)):
        pearson, p = pearsonr(data[ids[i]], data[ids[j]])
        if pearson == np.nan:
            continue
        pairwise[i][j] = pearson
```

```python
pairwise[5][3]
```

```python
# generate an initial estimate of the clusters by using a clustering 
# algorithm to group things with pearson > 0.9

initial_cluster = {}
members = []
cluster_count=0

# go through and find all matches > 90. Add them to sets
for i in range(len(ids)):
    # we need a cluster for this item if we don't have one
    if i not in initial_cluster:
        initial_cluster[i] = cluster_count
        members.append([i])
        cluster_count += 1
    for j in range(i+1, len(ids)):
        if j in initial_cluster:
            continue
        if pairwise[i][j] > 0.9:
            initial_cluster[j] = initial_cluster[i]
            members[initial_cluster[i]].append(j)
```

```python
# for any element in the list, I want to find if it matches another cluster better than its own cluster


def average_match(e, l):
    """The average match between element e and list l"""
    # filter out ourself as this will give us a nice 1.0 value
    # the problem with this is if we are a list of 1, anything will have a better value!
    l = [x for x in l if x != e]
    if len(l) == 0:
        return 0
    values = [pairwise[e][x] for x in l]
    return 1.0*sum(values)/len(values)

for i in range(len(ids)):
    best_score = 0
    best_id = None
    for j in range(len(members)):
        score = average_match(i, members[j])
        if score > best_score:
            best_score = score
            best_id = j
    if initial_cluster[i] != best_id:
        curr_score = average_match(i, members[initial_cluster[i]])
        print(str(i) + " current: " + str(initial_cluster[i]) + " best: " + str(best_id) + " current score: " + str(curr_score) + " best score: " + str(best_score))
        initial_cluster[i] = best_id
```

```python
centroids, _  = kmeans(np.array(pairwise), cluster_count)
idx, _ = vq(np.array(pairwise), centroids)
```
